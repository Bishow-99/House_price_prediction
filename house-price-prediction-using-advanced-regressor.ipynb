```python
# lets reduced the missing value to those columns which have less than 20 percent
def remove_null(dataset):
    null_percent=20
    min_count=int(((100-null_percent)/100)* dataset.shape[0])
    print(" The number of columns which have", min_count," are dropped")
    dataset.dropna(axis=1,thresh=min_count,inplace=True)
remove_null(train_data)
```

```python
# Now fill the missing value in both train and test dataset

def fill_value(df):
    for col in df.select_dtypes(exclude="object"):
        df[col].fillna(df[col].mean(), inplace=True)
    for col in df.select_dtypes(include="object"):
        df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
#Detect outlier using inter quartile range and change the feature of numerical dataset

no_outlier=[]
def remove_outlier(df):
    for col in df.columns:
        if col in train_num:
            q1=df[col].quantile(0.25)
            q3=df[col].quantile(0.75)
            iqr=q3-q1
            upper_bound=q3+1.5*iqr
            lower_bound=q1-1.5*iqr
            df[col]=np.where(df[col]> upper_bound,upper_bound,np.where(df[col]<lower_bound,lower_bound,df[col]))
            no_outlier.append(col)
    return no_outlier
remove_outlier(train_data)
        
```

```python
#update numerical feature in train dataset which have no outliers
train_num=train_data[no_outlier]
```

<h1 style="color:green; font-weight:bold; font-size:30px">Model Designing</h1>

```python
y=train_data['SalePrice']
X=train_data.drop("SalePrice", axis=1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2)
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,y_pred))
print(np.sqrt(mean_squared_error(y_test,y_pred)))
print(r2_score(y_test,y_pred))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
Lr=LinearRegression()
cv_value=5
parameters={'fit_intercept':[True, False],
            'copy_X':[True,False]
           }
    
    
linear_grid=GridSearchCV(Lr,param_grid=parameters,cv=cv_value,verbose=1,n_jobs=-1,)
linear_grid.fit(X_train,y_train)
y_pred=linear_grid.predict(X_test)
print("mean squared error",mean_squared_error(y_test,y_pred))
print("root mean squared error",np.sqrt(mean_squared_error(y_test,y_pred)))
print("r squared",r2_score(y_test,y_pred))
```

```python
print(linear_grid.best_score_)
print(linear_grid.best_params_)
```
